```python
from pyspark.sql import SparkSession

# Initialize SparkSession
spark = SparkSession.builder \
    .appName("Split TSV and Save to HDFS") \
    .getOrCreate()

# Read the TSV file into a DataFrame
input_file = "hdfs://192.168.2.246:9000/user/ubuntu/amazon_reviews_us_Baby_v1_00.tsv"
df = spark.read.option("delimiter", "\t").csv(input_file)

# Determine the desired size for each split file (e.g., in bytes)
split_size_bytes = 500 * 1024 * 1024  # Example: 100 MB

# Repartition the DataFrame into equal-sized partitions
df_repartitioned = df.repartitionByRange(int(df.count() * split_size_bytes / df.rdd.map(lambda x: len(x.encode('utf-8'))).sum()))

# Write each partition to a separate file in HDFS
output_dir = "hdfs://192.168.2.246:9000/user/ubuntu/half_baby.tsv"
df_repartitioned.write.option("delimiter", "\t").csv(output_dir)

# Stop SparkSession
spark.stop()

```
